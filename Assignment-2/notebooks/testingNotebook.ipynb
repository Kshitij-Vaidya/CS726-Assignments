```python
from sklearn.datasets import make_moons

# Make an image using the make_moons function from sklearn
def make_moons_image(n_samples: int = 1000, noise: float = 0.05, random_state: int = 0) -> Tuple[np.ndarray, np.ndarray]:
    X, y = make_moons(n_samples=n_samples, noise=noise, random_state=random_state)
    return X, y

X, y = make_moons_image(n_samples=1000, noise=0.05, random_state=0)
```

```python
# Plot the image
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue')
plt.show()
```
