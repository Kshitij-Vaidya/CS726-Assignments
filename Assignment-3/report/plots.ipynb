```python
import numpy as np
import matplotlib.pyplot as plt

# Data
methods = [
    "Greedy", "Random (τ=0.5)", "Random (τ=0.9)",
    "Top-k (k=5)", "Top-k (k=10)", "Nucleus (p=0.9)"
]
bleu_scores = [0.3099, 0.2928, 0.1985, 0.2476, 0.2490, 0.0846]
rouge_1_scores = [0.3543, 0.2932, 0.2000, 0.2464, 0.2584, 0.1473]
rouge_2_scores = [0.1296, 0.0931, 0.0585, 0.0698, 0.0797, 0.0422]
rouge_lcs_scores = [0.2710, 0.2266, 0.1647, 0.1826, 0.1984, 0.1153]

x = np.arange(len(methods))  # X-axis positions
width = 0.2  # Width of bars

# Plot BLEU scores
plt.figure(figsize=(12, 6))
plt.bar(x - 1.5 * width, bleu_scores, width, label="BLEU", color='b')
plt.bar(x - 0.5 * width, rouge_1_scores, width, label="ROUGE-1", color='g')
plt.bar(x + 0.5 * width, rouge_2_scores, width, label="ROUGE-2", color='r')
plt.bar(x + 1.5 * width, rouge_lcs_scores, width, label="ROUGE-LCS", color='y')

plt.xlabel("Decoding Methods")
plt.ylabel("Scores")
plt.title("Comparison of Decoding Methods")
plt.xticks(x, methods, rotation=20, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("decoding_methods.png")
# Show plot
plt.show()
```
